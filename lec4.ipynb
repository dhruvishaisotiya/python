```python
l=[1,2,3,4,5,6]
l1=[]
l2=[]
print(f"min{l[0]} max{l[-1]} min2{l[1]} max2{l[-2]}")
    
for i in l:
    if i%2==0:
        l1.append(i)
    else:
        l2.append(i)
print("even",l1)
print("odd",l2)
print("even max",max(l1))
print("odd max",max(l2))
print("even min",min(l1))
print("odd min",min(l2))
```
