```python

n=input("Enter password: ")
l=len(n)
u=0
o=0
d=0
s=0
e=0
if l>8:
    for i in n:
        if i.isupper():
            u=u+1
        elif i.islower():
            o=o+1
        elif i.isnumeric():
            d=d+1
        elif i.isspace():
            s=s+1
        else:
            e=e+1
    if u>0 and o>0 and d>0 and e>0 and s==0:
        print("valid password")
    else:
        print("invalid password")       
else:
    print("invalid password")
```

```python
n="xyz abc"
k=3
f=" "
n=n.upper()
ans="ABCDEFGHIJKLMNOPQRSTUVWXYZ"
for i in n:
    if i in ans:
        op=((ans.find(i)+k))%26
        f=f+ans[op]
    else:
        f=f+i
print(f)
```
