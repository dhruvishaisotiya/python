```python
# print count latter
s="hello how are you hello"
d={}
for i in s:
    d[i]=s.count(i)
d
```

```python
# print count word
x="hello how are you hello"
s=x.split(" ")
d={}
for i in s:
    d[i]=s.count(i)
d
```

```python
# first latter is same then count of word in list
x="dhruv vgh gdhh gh"
x=x.split()
d={}
for i in x:
    d[i[0]]=d.get(i[0],[])
    d[i[0]].append(i)
d
```

```python
# count latter
x="dhruv vgh gdhh gh"
d={}
for i in x:
    d[i]=d.get(i,0)+1
d
```
